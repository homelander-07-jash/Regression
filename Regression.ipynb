9. What is the main difference between Simple and Multiple Linear Regression ?

- SLR uses one predictor.

- MLR uses multiple predictors.

10.  What are the key assumptions of Multiple Linear Regression ?

- Linearity

- Independence

- Homoscedasticity

- Normality of residuals

- No or low multicollinearity.

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model ?

-  Non-constant variance of errors. It violates assumptions and may bias standard errors, affecting p-values.

12. How can you improve a Multiple Linear Regression model with high multicollinearity ?

- Remove or combine correlated variables.

- Use PCA or Ridge/Lasso regression.

- Check Variance Inflation Factor (VIF).

18. What are the limitations of using R² as a sole measure of model performance ?

- Doesn’t reflect overfitting

- Doesn’t measure predictive power on new data

- Always increases with more variables, even if irrelevant

19. How would you interpret a large standard error for a regression coefficient ?

- Indicates high uncertainty in the estimate → possibly due to multicollinearity, small sample size, or outliers.

20. How can heteroscedasticity be identified in residual plots, and why is it important to address it ?

- Plot residuals vs. predicted values

- Look for funnel shape or pattern → if present, it needs to be addressed (e.g., using transformations).

21.  What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R² ?

- ndicates irrelevant predictors are inflating R². Adjusted R² penalizes complexity, so a big drop suggests overfitting.

22. Why is it important to scale variables in Multiple Linear Regression?

- Ensures comparability

- Important for models with regularization

- Helps gradient-based optimization converge faster

31.  How is polynomial regression implemented in Python ?


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 5, 10, 17, 26])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit linear model on polynomial features
model = LinearRegression()
model.fit(X_poly, y)

# Predict and plot
y_pred = model.predict(X_poly)
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title("Polynomial Regression")
plt.show()
```
